## Getting the data

This time main goal of the project is to transform given data to graphable form to train principal component analysis.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

As we can see data has 30 atributes. If we would want to plot this, we would need 30 dimensions.
## Scaling data and transforming with PCA


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
print(scaled_data.shape)
print(x_pca.shape)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=cancer, x=x_pca[:,0],y=x_pca[:,1],hue='target')
plt.xlabel('First principal component')
plt.ylabel('Second Principal component')
```

as we have reduced the number of features, we cannot now clearly infer what is on the axes

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
plt.figure(figsize=(22,6))
sns.heatmap(df_comp,cmap='plasma',annot=True)
```

Above we can see heatmap with correlation of two created features with the original 30 features.
